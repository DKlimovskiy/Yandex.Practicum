<b> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ 2</b>
    
–ï—Å–ª–∏ –≤–º–µ—Å—Ç–æ –ø–æ–ª–∞ –≤–∑—è—Ç—å –≤–æ–∑—Ä–∞—Å—Ç–Ω—ã–µ –≥—Ä—É–ø–ø—ã, —Å–æ–∑–¥–∞–Ω–Ω—ã–µ –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ, —Ç–æ –∑–Ω–∞—á–µ–Ω–∏—è –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å –±–æ–ª–µ–µ —Ç–æ—á–Ω—ã–º–∏. –ù–æ, –ø–æ–∑–≤–æ–ª—å, —è –Ω–µ –±—É–¥—É —Å–æ–∑–¥–∞–≤–∞—Ç—å –∏—Ö :) –•–æ—á—É –ª–∏—à—å –∏–¥–µ—é –ø–æ–∫–∞–∑–∞—Ç—å. 

–ü—Ä–æ–ø—É—Å–∫–∏ —ç—Ç–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ. –í—ã–±–æ—Ä–∫–∞ –Ω–µ —Ç–∞–∫–∞—è –±–æ–ª—å—à–∞—è, –ø–æ—ç—Ç–æ–º—É –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Å–æ–¥–µ—Ä–∂–∞—Ç —Ç–æ–ª—å–∫–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏ –≤ —Ü–µ–ª–µ–≤–æ–º —Å—Ç–æ–ª–±—Ü–µ.</div>

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê

qq = df.pivot_table(index=['income_type', 'gender'], columns='education', values='total_income', aggfunc='median')
qq.loc[('—Å—Ç—É–¥–µ–Ω—Ç', 'M')]['–≤—ã—Å—à–µ–µ']
````

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê

def super_fillna_func(income_type, gender, education):
    '''
    –ù–∞—Ö–æ–¥–∏—Ç –≤ —Ç–∞–±–ª–∏—Ü–µ qq –Ω—É–∂–Ω—É—é –º–µ–¥–∏–∞–Ω—É.
    '''
    try:
        return qq.loc[(income_type, gender)][education]
    except:
        return 0
    
print(super_fillna_func('—Å—Ç—É–¥–µ–Ω—Ç', 'M','–≤—ã—Å—à–µ–µ') ) 
````

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê

print(super_fillna_func('—Ä–µ–≤—å—é–µ—Ä', 'F','–≤—ã—Å—à–µ–µ') )
````

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê

# –¢–∞–∫ —Å—Ä–∞–±–æ—Ç–∞–µ—Ç apply.
df.apply(lambda row: super_fillna_func(row['income_type'], row['gender'], row['education']), axis=1)
````

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê (–ù–µ –∑–∞–ø—É—Å–∫–∞—é)

# –í–æ—Ç —Ç–∞–∫ –ø—Ä–∏–º–µ–Ω—è–µ–º –∫ —Ç–∞–±–ª–∏—Ü–µ.

# –ó–∞–ø–∏—à–µ–º –≤ –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü.
df['new_income'] = df.apply(lambda row: super_fillna_func(row['income_type'], row['gender'], row['education']), axis=1)

# –ü–∞–Ω–¥–∞—Å —Å–∞–º –∑–∞–º–µ–Ω–∏—Ç –ø—Ä–æ–ø—É—Å–∫–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∏–∑ –Ω–æ–≤–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –≤ —Ç–æ–π –∂–µ —Å—Ç—Ä–æ–∫–µ.
df['total_income'] = df['total_income'].fillna(df['new_income'])
````

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê (–ù–µ –∑–∞–ø—É—Å–∫–∞—é)

# –ê –≤–æ—Ç —Ç–∞–∫ –º—ã —Å—Ç–æ–ª–±–µ—Ü —Å–æ–∑–¥–∞–¥–∏–º, –Ω–æ —Å–æ—Ö—Ä–∞–Ω—è—Ç—å –Ω–µ –±—É–¥–µ–º. –ó–∞–º–µ–Ω—É –ø—Ä–æ–∏–∑–≤–µ–¥–µ–º –≤ –≤–æ–∑–¥—É—Ö–µ:
df['total_income'] = df['total_income'].fillna(df.apply(lambda row: super_fillna_func(row['income_type'], row['gender'], row['education']), axis=1))
````

<div class="alert alert-success" style="border-radius: 15px; box-shadow: 4px 4px 4px; border: 1px solid ">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ 2 ‚úîÔ∏è</h2>
    
–°–ø–µ—à—É —Ç–µ–±—è –æ–±—Ä–∞–¥–æ–≤–∞—Ç—å. –ï—Å—Ç—å –µ—â–µ –±–æ–ª–µ–µ –∫–æ—Ä–æ—Ç–∫–∏–π –≤–∞—Ä–∏–∞–Ω—Ç: </div>

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê (–ù–µ –∑–∞–ø—É—Å–∫–∞—é)

df['total_income'] = df.groupby(['income_type','gender', 'education'])['total_income'].apply(lambda x: x.fillna(x.median()))
   
# –û–Ω–æ –∂–µ, –Ω–æ –±–µ–∑ –ª—è–º–±–¥—ã (—Ä–∞–±–æ—Ç–∞–µ—Ç –±—ã—Å—Ç—Ä–µ–µ):
df['total_income'] = df['total_income'].fillna(df.groupby(['income_type','gender', 'education'])['total_income'].transform('median'))
````

<div class="alert alert-success" style="border-radius: 15px; box-shadow: 4px 4px 4px; border: 1px solid ">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ 2 ‚úîÔ∏è</h2>
    
–≠—Ç–∏ –ø—Ä–∏–º–µ—Ä—ã –∫–∞–∂—É—Ç—Å—è —Ç—Ä—É–¥–Ω—ã–º–∏, –Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏ –∏—Ö —Å–µ–±–µ, –æ–Ω–∏ —Ç–µ–±–µ –≤ –±—É–¥—É—â–µ–º –º–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ —Å—ç–∫–æ–Ω–æ–º—è—Ç :) –õ—è–º–±–¥—ã —ç—Ç–æ —Ç–µ –∂–µ —Å–∞–º—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏, –Ω–æ –Ω–∞–ø–∏—Å–∞–Ω–Ω—ã–µ –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É. –ù–∏–∂–µ –ø—Ä–∏–º–µ—Ä:</div>

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê

# –°–æ–∑–¥–∞–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º.

l = {'A':[1,2,4], 'B':[3,3,5]}

dfq = pd.DataFrame(l)
dfq
````

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê

def sum_(row):
    ''' 
    –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ç–∞–±–ª–∏—Ü—É –ø–æ—Å—Ç—Ä–æ—á–Ω–æ. –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—É–º–º—É —ç–ª–µ–º–µ–Ω—Ç–æ–≤. 
    –ü—Ä–∏–º–µ–Ω—è–µ—Ç—Å—è –∫–æ –≤—Å–µ–π —Ç–∞–±–ª–∏—Ü–µ —á–µ—Ä–µ–∑ –º–µ—Ç–æ–¥ apply —Å –∞—Ä–≥—É–º–µ–Ω—Ç–æ–º axis=1, —Ç–æ –µ—Å—Ç—å –ø–æ—Å—Ç—Ä–æ—á–Ω–æ.
    '''
    return row['A']+row['B']

dfq['Sum'] = dfq.apply(sum_, axis=1)
dfq
````

<div style="border-radius: 15px; border: 3px solid black; padding: 10px;">
<b> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ 2</b>
    
–§—É–Ω–∫—Ü–∏—è –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É, —Ç–∞–∫ —á—Ç–æ –º–æ–∂–Ω–æ –ø—Ä–∏–º–µ–Ω–∏—Ç—å —Ç.–Ω. –ª—è–º–±–¥–∞-–≤—ã—Ä–∞–∂–µ–Ω–∏–µ. –≠—Ç–æ –±–µ–∑—ã–º—è–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è, –ø–æ—Ç–æ–º—É —á—Ç–æ –Ω–∞–º –Ω–µ –Ω—É–∂–Ω–æ –µ–µ –æ–±—ä—è–≤–ª—è—Ç—å, –≤–µ–¥—å –æ–Ω–∞ —É–º–µ—â–∞–µ—Ç—Å—è –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ—á–∫—É. –°–æ–≤–º–µ—â–∞–µ–º —ç—Ç–æ —Å apply, –∫–∞–∫ –∏ –ø—Ä–µ–∂–¥–µ. –ö—Å—Ç–∞—Ç–∏, apply —Ç–∞–∫ –∏ –ø–µ—Ä–µ–≤–æ–¥–∏—Ç—Å—è: –ø—Ä–∏–º–µ–Ω–∏—Ç—å.
</div>

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê

dfq['Sum2'] = dfq.apply(lambda row: row['A'] + row['B'], axis=1)
dfq
````

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê

# –°—É–º–º–∞ –ø–æ –≤—Å–µ–π —Å—Ç—Ä–æ–∫–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –≤–∑—è—Ç–∞ –ø—Ä–æ—â–µ:
dfq['Total_sum'] = dfq.apply(lambda row: row.sum(), axis=1)
dfq
````

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê

import numpy as np

# –ï—â–µ –ø—Ä–æ—â–µ, –Ω–æ —É–∂–µ –±–µ–∑ –ª—è–º–±–¥—ã. –°—É–º–º–∞ –≤—Å–µ–π —Å—Ç—Ä–æ–∫–∏:
dfq['Total_sum2'] = dfq.apply(np.sum, axis=1)
dfq
````

–¢–µ–ø–µ—Ä—å –ø–µ—Ä–µ–¥–∞–µ–º —Å—Ç–æ–ª–±—Ü—ã. –ê—Ä—É–≥–º–µ–Ω—Ç **axis=0**. </div>

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê

# –°—É–º–º–∞ —Å—Ç–æ–ª–±—Ü–∞. –î–æ–±–∞–≤–∏–º —Å—Ç—Ä–æ–∫—É.

dfq.loc[3] = dfq.apply(lambda col: col.sum(), axis=0)
dfq
````

````python
# –ö–û–î –†–ï–í–¨–Æ–ï–†–ê

# –°—É–º–º–∞ —Å—Ç–æ–ª–±—Ü–∞. –î–æ–±–∞–≤–∏–º —Å—Ç—Ä–æ–∫—É.
dfq.loc[4] = dfq.apply(np.sum, axis=0)
dfq
````

````python
def purpose_change(purpose):
    if '–∞–≤—Ç–æ' in purpose:
        return '–ø–æ–∫—É–ø–∫–∞ –∞–≤—Ç–æ–º–æ–±–∏–ª—è'
    elif '–∂–∏–ª—å' in purpose or '–Ω–µ–¥–≤' in purpose:
        return '–ø–æ–∫—É–ø–∫–∞ –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏'
    #elif '–Ω–µ–¥–≤' in purpose:
        #return '–ø–æ–∫—É–ø–∫–∞ –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏'
    elif '—Å–≤–∞–¥—å' in purpose:
        return '—Å–≤–∞–¥—å–±–∞'
    else:
        return '–æ–ø–ª–∞—Ç–∞ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è'
    
df['purpose'] = df['purpose'].apply(purpose_change)
df['purpose'].value_counts()
````

**–í—ã–≤–æ–¥**

### –ö–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö

–£–ø—Ä–æ—Å—Ç–∏–º –ø—Ä–æ–≤–µ—Ä–∫—É –≥–∏–ø–æ—Ç–µ–∑

–°–æ–∑–¥–∞–¥–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏—é —Å–µ–º—å–∏

````python
def family_category(row):
    if row == 0:
        return '–Ω–µ—Ç –¥–µ—Ç–µ–π'
    elif row < 3:
        return '–Ω–µ–±–æ–ª—å—à–∞—è —Å–µ–º—å—è'
    else:
        return '–º–Ω–æ–≥–æ–¥–µ—Ç–Ω–∞—è —Å–µ–º—å—è'
    
df['family_category'] = df['children'].apply(family_category)
df['family_category'].value_counts()
````

–î–∞–ª–µ–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∏—Ä—É–µ–º —É—Ä–æ–≤–µ–Ω—å –¥–æ—Ö–æ–¥–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤:


````python
def income_category(row):
    if row < 100000:
        return '–¥–æ—Ö–æ–¥ –¥–æ 100 —Ç—ã—Å—è—á'
    if row <= 150000:
        return '–¥–æ—Ö–æ–¥ –æ—Ç 100 –¥–æ 150 —Ç—ã—Å—è—á'
    if row <= 200000:
        return '–¥–æ—Ö–æ–¥ –æ—Ç 150 –¥–æ 200 —Ç—ã—Å—è—á'
    else:
        return '–¥–æ—Ö–æ–¥ —Å–≤—ã—à–µ 200 —Ç—ã—Å—è—á'
    
df['income_category'] = df['total_income'].apply(income_category)    
df['income_category'].value_counts()
````
